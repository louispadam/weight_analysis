```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import funcs.graphics as gph
import funcs.analysis as anl
```

```python
to_analyze = "data/weight_data_23.csv"

# Read in data
daily = pd.read_csv(to_analyze)
daily["Time"] = pd.to_datetime(daily["Time"])            # Convert 'Time' column to datetime

# Compute monthly statistics
monthly = anl.months_stats(daily)

# Initialize graphics objects
fig, ax = plt.subplots(figsize=(10, 5))

# Add daily and monthly data to the plot
ax = gph.days_plot(ax, daily)
ax = gph.months_plot(ax, monthly)

# Labels and formatting
ax.set_xlabel("Time")
ax.set_ylabel("Weight (lbs)")
ax.set_title("Weight Over Time with Monthly Averages")
ax.grid(
    axis="y",
    linestyle="--",
    linewidth=0.5
)

# Present figure
ax.legend()

plt.show()
```
